```python
# importing necessary libraries 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# setting dataframe from csv documnet
df = pd.read_csv("world-happiness-report-2021.csv")
```

```python
# displaying all regions established by the report with the number of countries belonging to them
df[['Regional indicator']].value_counts()
```

```python
# selecting smaller data units to work with
Western_Europe=df[df['Regional indicator']== 'Western Europe']
Central_and_Eastern_Europe=df[df['Regional indicator']== 'Central and Eastern Europe']
Europe = pd.concat([Western_Europe, Central_and_Eastern_Europe])
```

```python
# Complete results for first five rows
Western_Europe.head();
```

```python
Central_and_Eastern_Europe.head()
```

```python
# Visualization of the happiness index for Western Europe
plt.figure(figsize=(10,4))
sns.barplot(data=Western_Europe,x="Country name",y="Ladder score")
plt.xticks(rotation=90) 
plt.show()
```

- This plot is quite general, however it shows that least seven countries are the southernmost ones, also five of seven best scoring countries belong to the northernmost.

```python
# Visualization of the happiness index for Central and Eastern Europe
plt.figure(figsize=(10,4))
sns.barplot(data=Central_and_Eastern_Europe,x="Country name",y="Ladder score")
plt.xticks(rotation=90) 
plt.show()
```

- Also in Central and Eastern Europe, the least succesfull countries belong to southernmost, with one significant exception: Kosovo scoring third. How does this happened?

```python
# Trying to find connection between ladder score and gross domestic product
# Displaying parallel lineplots

plt.figure(figsize = (16,8))
plt.xticks(rotation=90)
sns.lineplot(x="Country name", y = "Logged GDP per capita", data = Western_Europe,marker='o',label='Western_Europe')
sns.lineplot(x="Country name", y = "Logged GDP per capita", data = Central_and_Eastern_Europe,marker='o',label='Central_and_Eastern_Europe')
plt.show()
```

- If this lineplots would be simplified into a straight line (which I failed to), it would most probably slowly decrease, showing relation between GDP and Happiness.
- However, the two countries with, by far, the highest GDP, Luxembourg and Ireland are not scoring the best in ladder score. Possible explanation: from certain level of "wealthiness", it has no significant effect to happiness.
- In Central and Eastern Europe, there are visible rebounds from the trend, by Kosovo, Serbia, Bosnia and Hercegovina and partly also by Albania and North Macedonia, Kosovo being the most significant. Possible explanation: All five countries had more or less significant experience with war in recent years.

```python
# Displaying another parallel lineplots showing generosity, where Kosovo is doing best
plt.figure(figsize = (16,8))
plt.xticks(rotation=90)
sns.lineplot(x="Country name", y = "Generosity", data = Western_Europe,marker='o',label='Western_Europe')
sns.lineplot(x="Country name", y = "Generosity", data = Central_and_Eastern_Europe,marker='o',label='Central_and_Eastern_Europe')
plt.show()
```

- These are very bouncy lineplots. All five countries mentioned above, score the best in term of generosity in Eastern and Central Europe.

```python
# selecting valid data in order to create a heatmap, that allow to observe data correlations
cols=['Country name', 'Ladder score','Logged GDP per capita', 'Social support', 
      'Healthy life expectancy', 'Freedom to make life choices', 'Generosity',
       'Perceptions of corruption']
```

```python
Western_Europe = Western_Europe[cols]

sns.heatmap(Western_Europe.corr(), vmin=-1, vmax=1,annot=True)
```

- In Western Europe, Social support and Freedom to make life choices have the srongest correlation to hapiness. However Logged GDP per capita and Generosity have strong correlation to Happines as well.
- There is also significant correlation between Logged GDP and Generosity.


```python
Central_and_Eastern_Europe = Central_and_Eastern_Europe[cols]

sns.heatmap(Central_and_Eastern_Europe.corr(), vmin=-1, vmax=1,annot=True)
```

- Central and Eastern Europe plot is quite different to Western Europe.
- Social support, GDP and Freedom to make life have correlation to Hapiness aswell, but not as strong as in Western Europe. Generosity has no correlation.
- Unlike Western Europe, GDP and Generosity are by far the most uncorrelated here.

```python
# setting scatterplot to find relationship between Generosity, GDP and Happiness again.
Western_Europe.plot.scatter(x="Generosity", y="Logged GDP per capita", c="Ladder score", figsize=(10,8))
```

- In Western Europe wealthier countries are generally more happy and generous too

```python
Central_and_Eastern_Europe.plot.scatter(x="Generosity", y="Logged GDP per capita", c="Ladder score", figsize=(10,8))
```

- Very different plot again. Countries with higher GDP appear to be less generous. 

```python
# setting boxplot to observe differences in generosity
Europe.boxplot(column=["Generosity"], by="Regional indicator", figsize=(10,6), grid=False)
```

- While generosity in Central and Eastern Europe is lower in general, there is also much smaller range, with the exception of Kosovo.

```python
# setting an interractive map
fig = px.choropleth(Europe,
                    scope = "europe",
                    locations="Country name",
                    locationmode='country names',
                    color='Ladder score', 
                    hover_name = 'Ladder score',
                    hover_data = cols,
                    color_continuous_scale = px.colors.sequential.Blues,
                   )

fig.update_layout(height=400, width=800, margin={"r":1,"t":0,"l":0, "b":0})
fig.show()
```
